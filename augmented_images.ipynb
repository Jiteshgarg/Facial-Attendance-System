```python
from PIL import Image
import os

# Directory containing the images
images_folder = 'images/'  # Replace 'your_folder_path' with the path to your folder

# List all image files in the directory
image_files = [f for f in os.listdir(images_folder) if os.path.isfile(os.path.join(images_folder, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg'))]
# Custom sorting function to extract the number after "image"
def sort_key(filename):
    return int(''.join(filter(str.isdigit, filename)))

# Sort the list based on the number after "image"
image_files= sorted(image_files, key=sort_key)

print(image_files)

# Initialize an empty list to store images
image_list = []

# Loop through each image file, open it, and append it to the list
for img_file in image_files:
    try:
        img_path = os.path.join(images_folder, img_file)
        img = Image.open(img_path)
        image_list.append(img)
    except Exception as e:
        print(f"Error reading {img_file}: {e}")

# Display the number of images read
print(f"Number of images read: {len(image_list)}")
```

```python
from PIL import Image, ImageEnhance
import random
augmented_images_list = []
# Function to perform image augmentation
def augment_images(image_list):
    

    for img in image_list:
        # Copy the image to avoid in-place modifications
        img_copy = img.copy()

        # Apply image augmentation techniques as before...
        # Example augmentation methods:
        enhancer_contrast = ImageEnhance.Contrast(img_copy)
        img_contrast = enhancer_contrast.enhance(random.uniform(0.5, 1.5))  # Adjust contrast randomly

        enhancer_brightness = ImageEnhance.Brightness(img_copy)
        img_brightness = enhancer_brightness.enhance(random.uniform(0.5, 1.5))  # Adjust brightness randomly

        img_right_tilt = img_copy.rotate(random.uniform(-15, 15), expand=True)
        # ... Other augmentations ...

        # Append augmented images to the list
        img_left_tilt = img.rotate(random.uniform(165, 195), expand=True)

    # Shadow adjustment
        img_shadow = ImageEnhance.Brightness(img).enhance(random.uniform(0.1, 0.9))

    # Zoom in and zoom out
        img_zoom_in = img.resize((int(img.width * random.uniform(1.1, 1.5)), int(img.height * random.uniform(1.1, 1.5))))
        augmented_images_list.append(img_copy)
        augmented_images_list.append(img_contrast)
        augmented_images_list.append(img_brightness)
        augmented_images_list.append(img_left_tilt)
        augmented_images_list.append(img_right_tilt)
        augmented_images_list.append(img_shadow)
        augmented_images_list.append(img_zoom_in)

    return

augment_images(image_list)
```

```python
import os

# Assuming you have a list of PIL Image objects named 'image_list'
# Create a folder to store augmented images
folder_name = 'augmented_images'
os.makedirs(folder_name, exist_ok=True)

# Save images to the folder
for i, img in enumerate(augmented_images_list):
    try:
        img.save(f"{folder_name}/augmented_image_{i+1}.jpg")
        print(f"Image {i+1} saved successfully.")
    except Exception as e:
        print(f"Error saving image {i+1}: {e}")
```

```python
email_list=df["Email Address"]
mail_id_train=[]
for mail in email_list:
    for i in range(7):
        mail_id_train.append(mail)
```

```python
import numpy as np
import tensorflow as tf
from keras.preprocessing import image

images = augmented_images_list
labels = mail_id_train

def preprocess_images(img_list):
    processed_imgs = []
    target_size = (299, 299) 
    for img in img_list:
        img = img.convert("RGB") 
        img_resized = img.resize(target_size) 
        img_array = image.img_to_array(img_resized)
        img_array = np.expand_dims(img_array, axis=0)
        processed_imgs.append((img_array))
    return np.vstack(processed_imgs)

from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D, Dropout
from keras.optimizers import Adam
from keras.applications import InceptionV3
import numpy as np
base_model = InceptionV3(weights='imagenet', include_top=False)
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.5)(x)  
x = Dense(512, activation='relu')(x)
predictions = Dense(len(set(labels)), activation='softmax')(x) 
model = Model(inputs=base_model.input, outputs=predictions)
for layer in base_model.layers:
    layer.trainable = False
custom_optimizer = Adam(lr=0.001)
model.compile(optimizer=custom_optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
processed_images = preprocess_images(images)
label_map = {label: idx for idx, label in enumerate(set(mail_id_train))}
labels_numeric = np.array([label_map[label] for label in labels])
model.fit(processed_images, labels_numeric, epochs=30, batch_size=32, validation_split=0.2)
model.save('face_recognition_model_2.h5')
```
